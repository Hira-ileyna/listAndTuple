```python
positions = [(2.5, 1.2), (5.4, 2.8), (3.1, -0.2) ,(4.9, 2.9)]
x_limits = (0.0, 5.0)
y_limits = (0.0, 3.0)

def is_safe_position(position, x_limits, y_limits):
    x, y = position
    x_min , x_max =x_limits
    y_min , y_max = y_limits

    if x_min <= x <= x_max and y_min <= y <= y_max:
      return True
    else:
      return False
for pos in positions:
  result = is_safe_position(pos, x_limits, y_limits)
  print(pos, result)
```

```python
currents = [12.0, 12.5, 13.0, 13.5, 14.0, 13.0]
def average_list(values):
  total = 0

  for v in values:
    total += v

  return total / len(values)

def count_above_limit(values, limit):
  count = 0

  for v in values:
    if v > limit:
      count += 1
  return count

def max_value(values):
  max_val = values[0]

  for v in values:
    if v > max_val:
      max_val = v

  return max_val

avg = average_list(currents)
count = count_above_limit(currents, 13.0)
maximum = max_value(currents)

print("Average: " , avg)
print("Above 13: ", count)
print("Max: " , maximum)
```

```python
samples = [(0, 20.1), (5, 20.4), (10, 21.0), (15, 23.6), (20, 22.2), (25, 19.7), (30, 20.3)]
limits = (20.0, 23.0)

def warning_times(samples, limits):
  low, high = limits
  warnings = []

  for minute, temp in samples:
    if temp < low or temp > high:
      warnings.append(minute)

  return warnings


def first_high_temprature_time(samples, high_limit):
  i = 0

  while i < len(samples):
    minute, temprature = samples[i]
    if temprature > high_limit:
      return minute
    i += 1

  return -1

warn = warning_times(samples, limits)
first_high = first_high_temprature_time(samples, limits[1])

print("Warning times: " , warn)
print("First high temprature time: " , first_high)
```
